```python
import pandas as pd
file_path = "C:\\Users\\zhi-n\\Documents\\FIT3163\\mimic-iv-note\\discharge.csv"
data = pd.read_csv(file_path)
patient_note = data['text'].head(1).iloc[0]  # one row
```

```python
import spacy
import en_core_sci_sm
from collections import Counter


nlp = en_core_sci_sm.load()

def preprocess_text(text):
    # Tokenize the text and apply preprocessing steps
    doc = nlp(text.lower())  # Convert text to lowercase
    processed_text = ' '.join([token.lemma_ for token in doc if not token.is_stop and not token.is_punct])
    return processed_text

def summarize_text(text, n_sentences=5):

    # Process the text to identify named entities and noun chunks
    doc = nlp(text)
    entities = set(doc.ents)  # Extract named entities
    noun_chunks = set(doc.noun_chunks)  # Extract noun chunks

    # Compute the importance of each sentence based on the presence of named entities and noun chunks
    sentence_importance = {}
    for sent in doc.sents:
        sentence_importance[sent] = 0
        for token in sent:
            if token in entities or any(token in chunk for chunk in noun_chunks):
                sentence_importance[sent] += 1

    # Normalize importance scores based on the length of the sentence
    for sent, importance in sentence_importance.items():
        sentence_importance[sent] = importance / len(sent)

    # Select the top n sentences with highest importance scores
    summarized_sentences = sorted(sentence_importance, key=sentence_importance.get, reverse=True)[:n_sentences]
    summary = ' '.join([sent.text for sent in summarized_sentences])
    return summary


# Processing
preprocessed_note = preprocess_text(patient_note)
summary = summarize_text(preprocessed_note, n_sentences=3)  

# Output
print("Summary:\n", summary)
```
